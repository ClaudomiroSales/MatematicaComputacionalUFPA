<font size="+1"> Inicialmente, são geradas aleatoriamente 50 sequências de três números (pontos flutuantes), no intervalo de -10 a 10. Abaixo um exemplo de duas sequência possíveis:  </font>

[ 5.5, -2.4, 9.3 ], [ -1.5, 8.7, -0.3 ]

<font size="+1">            </font>







<font size="+1"> Cada sequência é aplicada ao sistema acima nos respectivos x1, x2 e x3. Para cada equação, é avaliado percentualmente o quanto o lado esquerdo da equação se aproxima do lado direito. A média dos percentuais das três equações é o índice de avaliação para a sequência testada. Por exemplo, considere os percentuais parciais e finais da sequência abaixo: </font>

[ 5.5, -2.4, 9.3 ], 40% para equação 1, 70% para equação 2 e 70% para equação 3  => 
média = 60%


<font size="+1"> O objetivo do programa é maximizar o percentual médio iterativamente usando força bruta, por um processo conhecido como Random Walk  </font>

<font size="+1"> A avaliada cada sequência, elas são ordenadas da melhor para a pior. As 50% piores sequências são descartadas sendo substituídas por novas sequências aleatórias.  </font>

<font size="+1"> O programa deve parar quando é alcançado o número máximo de iterações ou o percentual de aproximação médio ultrapassar um valor máximo definido, como por exemplo, 99,9% ou 98,9%.  </font>

## Plot da média do percentual de aproximação das melhores sequências por iteração

<font size="+1"> O programa deve ser executado 5 vezes. Plotar o percentual médio das 5 execuções da melhor sequência por iteração.  </font>

```python
#percMelhor = np
numIterations = 1000
numExec = 5

percMelhor = np.zeros((numExec, numIterations ))
print(percMelhor)

print('\nComeça as 5 execuções.\n')


#Execução 0

print('Execução 0')
print('Iteração 0')

percMelhor[0 , 0] = 70

print(percMelhor)

print('Execução 0')
print('Iteração 1')

percMelhor[0 , 1] = 73

print('O processo continua até a última iteração')

print(percMelhor)


#Execução 1


print('Execução 1')
print('Iteração 0')

percMelhor[1 , 0] = 60

print(percMelhor)

print('Execução 1')
print('Iteração 1')

percMelhor[1 , 1] = 80

print('O processo continua até a última iteração')

print(percMelhor)

print('e assim por diante para as outras execuções')

#Para cada iteração calcular a média

#Calcular a média da melhor sequência por execução para cada iteração
meanMelhor = np.zeros(( numIterations ))
for iteration in np.arange(0, np.shape(percMelhor)[ 1 ]):
    meanMelhor[ iteration ] = np.mean( percMelhor[ :, iteration ] )
    
#Plot meanMelhor    
    
```
