<font size="+2">  <b>  4-  Considerando x e y para o caso 1, A fórmula para achar o erro relativo de s = (x + y) é ERs = ERx.( x’ / x’ + y’) + ERy.(y’ / y’ + x’) + RA. Mas como no caso 1 os valores x e y são exatos, a fórmula fica ERs = RA. Minha dúvida é sobre como achar esse RA.

Considerando x = 183,9, seu valor truncado é 0,183.10³ e arredondado 0,184.10³; e considerando y = 0,076, seu valor truncado é 0,076.10³ e 0,076.10³. RA resultaria em 3 possíveis resultados: a soma dos valores originais, a soma dos valores truncados e a soma dos valores arredondados. Deve ser considerado as 3 possibilidades? <b>  <font size="+2">

```python
trunc = np.absolute(0.1839*10**3 - 0.183*10**3 ) / ( 0.183*10**3 )
print(np.absolute(0.1839*10**3 - 0.183*10**3 ))
print('RA = trunc =', trunc)
print(trunc * 100, '%')
```
