# Comparação entre os métodos

Na seção anterior vimos vários métodos com o mesmo propósito, no entanto o único dado que os exemplos nos fornecem para medir a complexidade dos algoritmos é a quantidade de iterações efetuadas.

O método ideal é aquele em que a convergência estiver assegurada. O método de Newton é o mais indicado quando for fácil verificar as condições de convergencia e que o calculo de f'(x) não seja mto elaborado. Já quando é trabalho avaliar f'(x) é aconselhavel usar o método da secante pois é o que converge mais rapido dentre as outras opções.

No entanto temos que levar em consideração a condição de parada, se o objetivo for reduzir o intervalo que contém a raiz, o método da posição falsa e o da secante não devem ser usados. Besse caso Método do ponto fixo ou Newton seriam os que trabalham exclusivamente com aproximações Xk para a raiz exata.
