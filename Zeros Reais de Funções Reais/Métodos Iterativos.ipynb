# Método da Bissecção

Dada uma função contínua e com sinais contrários, onde f(a) e f(b) podem ser positivo e negativo respectivamente ou negativo e positivo, podemos calcular um ponto médio P sendo igual a a + b dividido por 2. Logo P=(a+b)/2.

Com isso podemos obter três resultados diferentes:
    
*  f(p) = 0 -> Encontramos a raiz da função
*  f(a) e f(P) possuem sinais contrátios -> A raiz se encontra no primeiro intervalo.
*  f(b) e f(P) possuem sinais contrátios -> A raiz se encontra no segundo intervalo.

O método da bissecção consiste em repetir esse processo repetidamente até encontrar uma raiz com uma precisão tão boa quanto a que queiramos.
